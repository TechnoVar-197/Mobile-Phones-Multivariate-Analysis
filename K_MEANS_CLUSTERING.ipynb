###Data Preparation
Data preparation follows a series of steps that starts with collecting the right data, followed by cleaning, labeling, and then validation and visualization.

```python
np.array(df["Screen Size (inches)"])
```

```python
df["Price ($)"]=df["Price ($)"].str.extract('(\d+)', expand=False)
```

```python
df["Price ($)"]
```

```python
df["Price ($)"]=df["Price ($)"].astype(int)
```

```python
df["RAM "]=df["RAM "].str.replace("GB","")
df["RAM "]=df["RAM "].astype(int)
```

```python
df["Storage "]=df["Storage "].str.replace("GB","")
df["Storage "]=df["Storage "].astype(int)
```

```python
df['n_cameras'] = df['Camera (MP)'].str.count('\\+') + 1
```

```python
res1 = []
res2 = []
res3 = []
res4 = []
for x in df['Camera (MP)']:
    resolutions = x.split('+')
    tam = len(resolutions)

    if tam == 1:
        res1.append(resolutions[0])
        res2.append('0')
        res3.append('0')
        res4.append('0')

    if tam == 2:
        res1.append(resolutions[0])
        res2.append(resolutions[1])
        res3.append('0')
        res4.append('0')

    if tam == 3:
        res1.append(resolutions[0])
        res2.append(resolutions[1])
        res3.append(resolutions[2])
        res4.append('0')

    if tam == 4:
        res1.append(resolutions[0])
        res2.append(resolutions[1])
        res3.append(resolutions[2])
        res4.append(resolutions[3])

df['res1'] = res1
df['res2'] = res2
df['res3'] = res3
df['res4'] = res4

df= df.drop(columns='Camera (MP)')
df.head()
```

```python
df['Screen Size (inches)'].replace(regex=True, inplace=True, to_replace=r'[^0-9.\-]', value=r'')
```

```python
cem1 = []
cem2 = []
cem3 = []
for x in df['Screen Size (inches)']:
    resolutions = x.split('.')
    tam = len(resolutions)

    if tam == 1:
        cem1.append(resolutions[0])
        cem2.append('0')
        cem3.append('0')

    if tam == 2:
        cem1.append(resolutions[0])
        cem2.append(resolutions[1])
        cem3.append('0')

    if tam == 3:
        cem1.append(resolutions[0])
        cem2.append(resolutions[1])
        cem3.append(resolutions[2])


df['cem1'] = cem1
df['cem2'] = cem2
df['cem3'] = cem3

df= df.drop(columns='Screen Size (inches)')
```

```python
df["screen"] = df['cem1']+"."+ df["cem2"]
df=df.drop(["cem1","cem2","cem3"],axis=1)
df
```

###K-MEANS CLUSTERING

```python
plt.scatter(df['RAM '],df['Price ($)'])
```

```python
km=KMeans(n_clusters=3)
```

```python
km
```

```python
y_predicted = km.fit_predict(df[['RAM ','Price ($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
```

```python
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
```

```python
plt.scatter(df1['RAM '],df1['Price ($)'],color='green')
plt.scatter(df2['RAM '],df2['Price ($)'],color='red')
plt.scatter(df3['RAM '],df3['Price ($)'],color='black')
plt.xlabel('RAM ')
plt.ylabel('Price ($)')
plt.legend()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Price ($)']])
df['Price ($)']=scaler.transform(df['Price ($)'].values.reshape(-1,1))
df
scaler.fit(df[['RAM ']])
df.SepalLengthCm = scaler.transform(df['RAM '].values.reshape(-1,1))
df
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['RAM ','Price ($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
```

```python
plt.scatter(df1['RAM '],df1['Price ($)'],color='green')
plt.scatter(df2['RAM '],df2['Price ($)'],color='red')
plt.scatter(df3['RAM '],df3['Price ($)'],color='black')
plt.xlabel('RAM ')
plt.ylabel('Price ($)')
plt.legend()
```

```python
k_rng= range(1,10)
sse = []
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df[['RAM ','Price ($)']])
  sse.append(km.inertia_)

sse
```

```python
plt.xlabel('k')
plt.ylabel('SSE')
plt.plot(k_rng,sse)
```
